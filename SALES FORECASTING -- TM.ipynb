```python
#pivot table basically here finding the average of all the different food types taken from item
#identifier column
#DRA12 are food types and 11.6 are the averages.


item_weight_mean = file.pivot_table(values = "Item_Weight", index = 'Item_Identifier')
display (item_weight_mean)
```

```python
display (file['Item_Identifier'])
```

```python
#missing values are getting filled as per their average are been taken above
#enumerate() allows us to iterate through a sequence 
#but it keeps track of both the index and the element

for i, item in enumerate(file['Item_Identifier']):
    if miss_bool[i]:
        if item in item_weight_mean:
            file['Item_Weight'][i] = item_weight_mean.loc[item]['Item_Weight']
        else:
            file['Item_Weight'][i] = np.mean(file['Item_Weight'])
```

```python
#outlet identifier column needs label encoding and i assigned it into a new column.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
file['Outlet'] = le.fit_transform(file['Outlet_Identifier'])
display (file['Outlet'])
```

```python
#one hot encoding for different distinct values.

file = pd.get_dummies(file, columns=['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'New_Item_Type'])
display (file.head())
```

```python
#dropping unnecessary columns

X = file.drop(columns=['Outlet_Establishment_Year', 'Item_Identifier', 'Outlet_Identifier', 'Item_Outlet_Sales'])
X.head()
```

```python
y = file['Item_Outlet_Sales']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print (X_train.shape, X_test.shape ,  y_train.shape, y_test.shape)
```

calling a function below and then we will call it everytime we will use any other models..

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
def train(model, X, y):
    # training the model
    model.fit(X, y)
    
    pred = model.predict(X)
    # perform cross-validation
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("CV Score:", cv_score)
    print("R2_Score:", r2_score(y,pred))
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
model = LinearRegression(normalize=True)
train(model, X_train, y_train)
coef = pd.Series(model.coef_, X.columns).sort_values()
print (coef)
coef.plot(kind='bar', title="Model Coefficients")
plt.show()
```

```python
model = Ridge(normalize=True)
train(model, X_train, y_train)
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind='bar', title="Model Coefficients")
plt.show()
```

```python
model = Lasso()
train(model, X_train, y_train)
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind='bar', title="Model Coefficients")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model,X_train, y_train)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, X_train, y_train)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
train(model, X_train, y_train)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
plt.show()
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor()
train(model, X_train, y_train)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
plt.show()
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()
train(model, X_train, y_train)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance")
plt.show()
```
